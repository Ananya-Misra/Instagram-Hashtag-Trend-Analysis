```python

from selenium import webdriver
import time
from webdriver_manager.chrome import ChromeDriverManager
import sqlite3
from selenium.webdriver.common.by import By
```

```python

mdriver = webdriver.Chrome(ChromeDriverManager().install())


def readSqliteTable():
    # try:
    sqliteConnection = sqlite3.connect('hashtags.sqlite3')
    cursor = sqliteConnection.cursor()
    print("Connected to SQLite")

    sqlite_select_query = "SELECT * from hashtag3 ORDER BY tag ASC"
    cursor.execute(sqlite_select_query)
    records = cursor.fetchall()
    sqliteConnection.close()
    print("Total rows are:  ", len(records))
    print("Printing each row")
    allqueries = ""
    for row in records:
        # print("Id: ", row[0])
        # print("Tag: ", row[1])
        # print("link: ", row[2])
        # print("page: ", row[3])
        # print("date: ", row[4])
        # print()
        try:
            val = tag_data(mdriver, row)
            sql_update_query = f"""Update hashtag3 set posts = '{val}' where link= '{row[2]}';"""
            allqueries += sql_update_query
            print(sql_update_query)
        except Exception as e:
            print(e)
    conn = sqlite3.connect('hashtags.sqlite3')
    cursor = conn.cursor()
    cursor.executescript(allqueries)
    conn.commit()
    conn.close()


# except sqlite3.Error as error:
#     print("Failed to read data from sqlite table", error)
# finally:
#     if sqliteConnection:
#         sqliteConnection.close()
#         print("The SQLite connection is closed")

readSqliteTable()
```

```python
def tag_data(driver, row):
    time.sleep(3)
    prefix = 'https://www.instagram.com/' + row[2]
    driver.get(prefix)
    time.sleep(3)
    posts = driver.find_element(By.CLASS_NAME, "_ac2a")
    print(posts.text)
    return posts.text





```
